```
class Matrix:
  def __init__(self,n,m,values):
    self.n=n
    self.m=m
    self.rows=values
    if self.rows==[0]:
      self.rows=list() 
      y=[0]*self.n 
      for i in range(self.m):
        self.rows.append(y)
    else:
      self.rows=list()
      if len(values)==self.m:
        for i in values:
          if len(i)==self.n:
            self.rows.append(i)
          else: 
            print('Error: # of columns does not match indicated column length n')
            break
      else: print('Error: # of rows does not math indicated row length m')

  def showmatrix(self):
    return self.rows

  def printmatrix(self):
    for i in self.rows:
      print(i)
 
  def __getitem__(self,M):
    return self.rows[M]

  def __setitem__(self,M_1,M_2):
    self.rows[M_1] = M_2

  def shape(self):
    return (self.n,self.m)

  def transpose(self):
    q=list(zip(*self.rows))
    return Matrix(self.m,self.n,q)

  def row(self,n):
    for i in range(len(self.rows)+1):
      if i==n:
        return self.rows[i-1]
      else: continue

  def to_list(self):
    return self.rows

  def scalarmul(self,c):
    for x in range(self.m):
      for y in range(self.n):
        self.rows[x][y]=c*(self.rows[x][y])
    return self.rows

  def add(self,N):
    mat= Matrix(self.n,self.m,[0])

    for x in range(self.m):
      for y in range(self.n):
        mat[x][y]=self.rows[x][y]+N[x][y]
    return mat

  def sub(self,N):
    tam= Matrix(self.n,self.m,[0])

    for x in range(self.m):
      for y in range(self.n):
        tam[x][y]=self.rows[x][y]-N[x][y]
    return tam

  def equals(self,N):
    for x in range(self.m):
      for y in range(self.n):
        while self.rows[x][y]==N[x][y]:
          return True
        else: break
    return False
    
    
    
    
    

 
```

```
def constant(n,m,c):
  list_1=list()
  for i in range(m):
    y=[c]*n
    list_1.append(y)
  return list_1

def zeros(n,m):
  list_2=list()
  for i in range(m):
    y=[0]*n
    list_2.append(y)
  return list_2

def ones(n,m):
  list_3=list()
  for i in range(m):
    y=[1]*n
    list_3.append(y)
  return list_3

def eye(n):
  list_4=list()
  for i in range(n):
    y=[0]*n
    list_4.append(y)
    list_4[i][i]=1
  return list_4
```
